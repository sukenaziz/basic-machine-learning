```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('seaborn')
```

```python
df_iris = pd.read_csv('dataset/iris.data', names=['sepal_length',
                                                 'sepal_width',
                                                 'petal_length',
                                                 'petal_width',
                                                 'species'])
df_iris.head()
```

```python
print('Total Rows:', df_iris.shape[0])
print('Total Column:', df_iris.shape[1])
```

```python
df_iris['species'].unique()
```

```python
df_iris.describe()
```

```python
sns.distplot(df_iris['sepal_width'])
plt.show()
```

```python
df_iris['species'].hist()
plt.show()
```

```python
df_iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
plt.show()
```

```python
df_iris.plot(kind='scatter', x='petal_length', y='petal_width')
plt.show()
```

```python
sns.boxplot(data=df_iris)
plt.xlabel('Features')
plt.ylabel('Centimeter')
plt.show()
```

```python
sns.boxplot(data=df_iris, x='species', y='petal_length')
plt.xlabel('Species')
plt.ylabel('Centimeter')
plt.show()
```

#### Pairplot

```python
sns.pairplot(data=df_iris, hue='species')
plt.show();
```

#### Heatmap (annottion for display the value on plot)

```python
sns.heatmap(data=df_iris.corr(), annot=True)
plt.show()
```
