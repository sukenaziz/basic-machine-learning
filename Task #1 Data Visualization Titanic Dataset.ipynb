```python
import sys
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df_titanic=pd.read_csv('dataset/titanic.csv')
df_titanic
```

```python
new_columns = {'Survived': 'SURVIVED', 
 'Pclass': 'PCLASS', 
 'Name': 'NAME', 
 'Sex': 'SEX', 
 'Age': 'AGE', 
 'Siblings/Spouses Aboard': 'SIBSA', 
 'Parents/Children Aboard': 'PARCA' , 
 'Fare': 'FARE'}
df_titanic = df_titanic.rename(columns=new_columns)
df_titanic
```

## Single Variable Plot

```python
sb.set_style(style='darkgrid')
sb.histplot(data=df_titanic, x='AGE')
```

## Two Variables Plot

```python
sb.scatterplot(data=df_titanic, x='AGE', y='FARE')
```

## Three Variables Plot

```python
sb.scatterplot(data=df_titanic, x='AGE', y='FARE',hue='PCLASS',palette='Dark2')
```

## Multi Variables Plot

```python
sb.heatmap(data=df_titanic.corr(), annot=True)
```
